```python
import os
import torch
import clip
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset
from PIL import Image
import torch.nn as nn
import torch.optim as optim

# Device setup
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load the CLIP model
clip_model, preprocess = clip.load("ViT-B/32", device=device)

# Image transformations (CLIP preprocess pipeline)
transform = preprocess


class ImageDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        """
        Args:
            root_dir (str): Path to dataset containing subfolders for each sentiment.
            transform (callable, optional): Transform to apply to images.
        """
        self.image_paths = []
        self.labels = []
        self.transform = transform

        # Fixed label mapping
        self.label_mapping = {'happy': 0, 'sad': 1, 'anger': 2}

        # Read image paths and labels
        for label in os.listdir(root_dir):
            label_dir = os.path.join(root_dir, label)
            if os.path.isdir(label_dir):
                for image_file in os.listdir(label_dir):
                    self.image_paths.append(os.path.join(label_dir, image_file))
                    self.labels.append(self.label_mapping[label])

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        label = self.labels[idx]

        # Load and preprocess the image
        image = Image.open(image_path).convert("RGB")
        if self.transform:
            image = self.transform(image)

        return image, label


class CLIPSentimentModel(nn.Module):
    def __init__(self, clip_embedding_dim, hidden_dim, num_classes):
        super(CLIPSentimentModel, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(clip_embedding_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, num_classes)
        )

    def forward(self, x):
        return self.fc(x)


def train_clip_model(train_loader, test_loader, num_classes=3, num_epochs=10, lr=1e-4):
    model = CLIPSentimentModel(clip_embedding_dim=512, hidden_dim=256, num_classes=num_classes).to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    for epoch in range(num_epochs):
        model.train()
        total_loss = 0.0
        correct = 0
        total = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)

            # Extract CLIP image embeddings
            with torch.no_grad():
                image_features = clip_model.encode_image(images)

            # Forward pass
            outputs = model(image_features)
            loss = criterion(outputs, labels)

            # Backward pass
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Metrics
            total_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        train_accuracy = correct / total
        print(f"Epoch {epoch + 1}/{num_epochs}, Loss: {total_loss / len(train_loader):.4f}, Accuracy: {train_accuracy:.4f}")

        # Validate the model
        validate_clip_model(model, test_loader, num_classes)

    # Save the trained model
    torch.save(model.state_dict(), "D:/SML Project/SML_Dataset/data/CLIP/clip_sentiment_model.pth")
    print("Model saved as clip_sentiment_model.pth")


def validate_clip_model(model, test_loader, num_classes=3):
    model.eval()
    correct = 0
    total = 0
    class_correct = [0] * num_classes
    class_total = [0] * num_classes

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)

            # Extract CLIP image embeddings
            image_features = clip_model.encode_image(images)

            # Forward pass
            outputs = model(image_features)
            _, preds = torch.max(outputs, 1)

            correct += (preds == labels).sum().item()
            total += labels.size(0)

            for i in range(len(labels)):
                label = labels[i].item()
                class_total[label] += 1
                if preds[i] == labels[i]:
                    class_correct[label] += 1

    # Overall accuracy
    accuracy = correct / total
    print(f"Overall Test Accuracy: {accuracy:.4f}")

    # Per-class accuracy
    for i in range(num_classes):
        if class_total[i] > 0:
            print(f"Class {i} Accuracy ({['happy', 'sad', 'anger'][i]}): {class_correct[i] / class_total[i]:.4f}")
        else:
            print(f"Class {i} has no samples.")


if __name__ == "__main__":
    train_dir = r'D:\SML Project\SML_Dataset\data\train'
    test_dir = r'D:\SML Project\SML_Dataset\data\test'

    train_dataset = ImageDataset(train_dir, transform=transform)
    test_dataset = ImageDataset(test_dir, transform=transform)

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

    train_clip_model(train_loader, test_loader, num_classes=3, num_epochs=10, lr=1e-4)
```
